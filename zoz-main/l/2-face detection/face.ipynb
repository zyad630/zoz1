```python
import cv2
from ultralytics import YOLO
model = YOLO(r"C:\Users\ITC\Downloads\yolov8n.pt")
cap =cv2.VideoCapture(1)
while True:
    ret,frame=cap.read()
    if not ret:
        print("afta77 alcameraaa")
        break
    result = model(frame)
    frame2 = result[0].plot()
    cv2.imshow('camera video',frame2)
    if cv2.waitKey(1)==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
